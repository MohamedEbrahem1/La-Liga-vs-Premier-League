# The Competitive Gap between Premier League and LaLiga - statistical analysis on the difference between the two leagues


## Import Libraries used

```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import numpy as np
```

## Import datasets used, the used data set are for both leagues regarding 2018-2019 season

```python
sp = pd.read_csv("H:\My Drive\Spring 2023\RHET 1020\Data analysis\season-1819_csv.csv")
pl = pd.read_csv("https://datahub.io/sports-data/english-premier-league/r/season-1819.csv")
```

```python
sp.sample(10)
```

## HomeTeam win odds in both leagues

```python
x = sp.groupby('HomeTeam')['B365H'].mean()
y = pl.groupby('HomeTeam')['B365H'].mean()
plt.plot(x,'b.')
plt.plot(y,'r.')
plt.legend(['LaLiga','Premier League'])
plt.title('Home Team  win odds')
plt.xticks(rotation = 'vertical')
plt.show()
```

## Draw odds in both leagues

```python
figsize = (6,200)
web = 'B365'
x = sp.groupby('HomeTeam')[web+'D'].mean()
y = pl.groupby('HomeTeam')[web+'D'].mean()
plt.plot(x,'b.')
plt.plot(y,'r.')
plt.legend(['LaLiga','Premier League'])
plt.title('Draw Odds')
plt.xticks(rotation = 'vertical')
plt.show()
```

## AwayTeam win odds in both leagues

```python
figsize = (6,200)
x = sp.groupby('AwayTeam')[web+'A'].mean()
y = pl.groupby('AwayTeam')[web+'A'].mean()
plt.plot(x,'b.')
plt.plot(y,'r.')
plt.legend(['LaLiga','Premier League'])
plt.title('Away Team Win odds')
plt.xticks(rotation = 'vertical')
plt.show()
```

```python
web='B365'
index = np.zeros(20)
for i in range(20):
    index[i] = i
x = pd.DataFrame(
    {"a" : sp['HomeTeam'].unique(),
    "b" : np.zeros(20)}, index)
for i in range(0,len(sp)):
    if (sp[web+'H'][i]>sp[web+'A'][i] and sp['FTR'][i] == 'H') or (sp[web+'H'][i]<sp[web+'A'][i] and sp['FTR'][i] == 'A'):
        f = sp['HomeTeam'][i]
        x.loc[x['a'] == f,['b']] += 1
        f = sp['AwayTeam'][i]
        x.loc[x['a'] == f,['b']] += 1
x1 = plt.bar(x['a'],x['b'])
# plt.xticks(x['a'], rotation = 'vertical')
# Time for the premier league
index = np.zeros(20)
for i in range(20):
    index[i] = i
y = pd.DataFrame(
    {"a" : pl['HomeTeam'].unique(),
    "b" : np.zeros(20)}, index)
for i in range(0,len(pl)):
    if (pl[web+'H'][i]>pl[web+'A'][i] and pl['FTR'][i] == 'H') or (pl[web+'H'][i]<pl[web+'A'][i] and pl['FTR'][i] == 'A'):
        f = pl['HomeTeam'][i]
        y.loc[y['a'] == f,['b']] += 1
        f = pl['AwayTeam'][i]
        y.loc[y['a'] == f,['b']] += 1
plt.bar(y['a'],y['b'])
plt.xticks([])
plt.legend(['LaLiga','Premier League'])
plt.title("Number of unexpected results in LaLiga vs Premier League by teams 2018-2019 season")
```
