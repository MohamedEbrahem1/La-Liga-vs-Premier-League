# For the analysis of the unexpected matches, we have to methods to calculate it:

1. See whether the winning side on the bet website was different from the real winning side (betting website method)
2. See whether the winning side have changed or not after the half time (half-time method)

## Betting website method

```python
# in this block, we initialize a variable called web to have the value of the betting website
web = bet
# calculate the number of teams in the first league
teams1 = sp['HomeTeam'].nunique()
index = np.zeros(teams1)
for i in range(teams1):
    index[i] = i
x = pd.DataFrame(
    {"a" : sp['HomeTeam'].unique(),
    "b" : np.zeros(teams1)}, index)
for i in range(0,len(sp)):
    if (sp[web+'H'][i]>sp[web+'A'][i] and sp['FTR'][i] == 'H') or (sp[web+'H'][i]<sp[web+'A'][i] and sp['FTR'][i] == 'A'):
        f = sp['HomeTeam'][i]
        x.loc[x['a'] == f,['b']] += 1
        f = sp['AwayTeam'][i]
        x.loc[x['a'] == f,['b']] += 1
x1 = plt.bar(x['a'],x['b'])
# Time for the Second league
#Calculate the number of teams in the second League
teams2 = pl['HomeTeam'].nunique()
index = np.zeros(teams2)
for i in range(teams2):
    index[i] = i
y = pd.DataFrame(
    {"a" : pl['HomeTeam'].unique(),
    "b" : np.zeros(teams2)}, index)
for i in range(0,len(pl)):
    if (pl[web+'H'][i]>pl[web+'A'][i] and pl['FTR'][i] == 'H') or (pl[web+'H'][i]<pl[web+'A'][i] and pl['FTR'][i] == 'A'):
        f = pl['HomeTeam'][i]
        y.loc[y['a'] == f,['b']] += 1
        f = pl['AwayTeam'][i]
        y.loc[y['a'] == f,['b']] += 1
plt.bar(y['a'],y['b'])
plt.xticks([])
plt.legend([my_dict[league1],my_dict[league2]])
plt.title("Unexpected results using the betting method")
```

## Half-time method

```python
x = 0
for i in range(0,len(sp)):
    if (sp['FTR'][i] != sp['HTR'][i]):
        x = x + 1
        plt.plot(sp['HomeTeam'][i],sp['AwayTeam'][i],'.')
        plt.xticks(sp['HomeTeam'],rotation = "vertical")
plt.title("Unexpected matches in " + my_dict[league1] + " in terms of different half-time and final results")
plt.show()
print("The number of unexpected matches using the changing of the result in half time method for " + my_dict[league1] + " is " + str(x))


x = 0
for i in range(0,len(pl)):
    if (pl['FTR'][i] != pl['HTR'][i]):
        x = x + 1
        plt.plot(pl['HomeTeam'][i],pl['AwayTeam'][i],'.')
        plt.xticks(pl['HomeTeam'],rotation = "vertical")
plt.title("Unexpected matches in " + my_dict[league2] + " in terms of different half-time and final results")
plt.show()
print("The number of unexpected matches using the changing of the result in half time method for " + my_dict[league2] + " is " + str(x))
```
